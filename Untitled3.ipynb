```python
import spacy 
from spacy import displacy
```

```python
nlp = spacy.load('en')
```

# NER - Named Entity Recognition

```python
doc = nlp("My name is John Carter.I live in Jakarta and work for the company Micorsoft Inc and on 12 June, 2020 i am doing good.")
```

```python
displacy.render(doc,style='ent',jupyter=True)
```
