# Linear Regression using numpy

<img src="images/linear_regression_1.png">

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#area
x = [1.2,2.6,3.2,1.8,2.5,3.6,4.2,6.8,6.4,7.2,8.2,9.1,8.2,4.2,5.6,3.9,1.5]
#price
y = [180,290,360,170,230,320,390,580,650,690,790,850,860,470,620,420,190]

x = np.array(x)
y = np.array(y)
```

<img src="images/linear_regression-2.png">

```python
# random initialization of m and c
m = 10
c = 20
learning_rate = 0.03

for i in range(500):
    yhat = x*m+c
    E = np.mean(np.square(y-yhat))
    m = m - learning_rate*2*np.mean((yhat-y)*x)
    c = c - learning_rate*2*np.mean(yhat-y)
    print(E)
    
    
```

```python
print(m)
print(c)
```

```python
def predict(x):
    yhat = m*x+c
    return yhat
```

```python
ypred = predict(x)
plt.scatter(x,y)
plt.plot(x,ypred,c='r')
plt.show()
```
